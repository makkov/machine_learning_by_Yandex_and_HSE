1. Загрузите данные close_prices.csv. В этом файле приведены цены акций 30 компаний на закрытии торгов за каждый день периода.
2. На загруженных данных обучите преобразование PCA с числом компоненты равным 10. Скольких компонент хватит, чтобы объяснить 90% дисперсии?
3. Примените построенное преобразование к исходным данным и возьмите значения первой компоненты.
4. Загрузите информацию об индексе Доу-Джонса из файла djia_index.csv. Чему равна корреляция Пирсона между первой компонентой и индексом Доу-Джонса?
5. Какая компания имеет наибольший вес в первой компоненте? Укажите ее название с большой буквы.


Загрузим данные, посмотрим на них

```python
import pandas
data = pandas.read_csv('close_prices.csv')
```

```python
data.head()
```

Уберем признак "дата"

```python
dataX = data.loc[:, 'AXP':]
```

Обучим преобразование PCA

```python
from sklearn.decomposition import PCA

X = dataX
pca = PCA(n_components=10)
pca.fit(X)
```

Определим, скольких компонент хватит, чтобы объяснить 90% дисперсии

```python
comp = 0
countComp = 0
for v in pca.explained_variance_ratio_:
    countComp += 1
    comp += v
    if comp >= 0.9:
        break

print countComp
```

Применим преобразование к исходным данным

```python
dataNew = pandas.DataFrame(pca.transform(X))
zeroComp = dataNew[0]
```

Загрузим информацию об индексе Доу-Джонса

```python
dataDjiaIndex = pandas.read_csv('djia_index.csv')
```

```python
dataDjiaIndex.head()
```

```python
djiaIndex = dataDjiaIndex['^DJI']
```

Определим корреляцию Пирсона между первой компонентой и индексом Доу-Джонса

```python
from numpy import corrcoef

corrcoef(zeroComp, djiaIndex)
```

Определим компанию с наибольшим весом в первой компоненте

```python
company = pandas.Series(pca.components_[0])
topCompany = company.sort_values(ascending=False).head(1).index[0]
answer = X.columns[topCompany]
```
